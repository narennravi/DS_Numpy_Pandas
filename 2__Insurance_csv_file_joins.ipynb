# **How to apply conditional filtering..?**

```python
df.head()
```

## 1. **Single** Condition

```python
# applying "single condition"
df[df["age"] > 40]
```

## 2. **Multiple** condition


*   And operator
*   Or operator



```python
# applying "multiple condition"
# and operator --> &

df[(df["sex"] == "male") & (df["smoker"] == "yes") & (df["age"] > 40) & (df["children"] > 2)]

#NOTE:
# All condition must satisfy in AND operator 
# Perfect filter of result
```

```python
# applying the multiple condition
# or operator ---> |

df[(df["sex"] == "male") | (df["smoker"] == "yes") | (df["age"] > 40) | (df["children"] > 2)]

#NOTE:
# Any 1 condition satisfy also consider and displayed in OR operator
# Least Perfect of result
```

# Example: 2

## How to add Column and **calculate expenses which got reduced by 10%** ..?

Expenses is reduced to 10% from 100% so difference is 90.

>  = 90/100 

> = 9/10

> = 0.9

reduced exps = Exps * 0.9





```python
df["reduced_expense"] = df["expenses"]*0.9
df.head()

#NOTE:
# new column --> "reduced expenses" is added with calculation of reduced 10%
```

```python
df["difference"] = df["expenses"] - df["reduced_expense"]
df.head()

#NOTE:
# Display the difference of reduced expense and expenses
```

## **Dropping column**
1. **Without inplace** function in Df --> previous unupdated value will displayed 

```python
df.drop(["difference","expenses"] , axis = "columns")

#NOTE:
# Here differences & expenses column is dropped
# axis is mentioning whether column (or) rows for dropping
```

```python
df.head()

#NOTE:
# After dropping the 2-column(difference,expense) it display asusual file bcz., it only executes expression for that execution time
# Difference , expense --> displayed. Bcz changes[Not updated in df]
```

2. **With inplace function** in df - updated vale got displayed

```python
df.drop(["difference","expenses"] , axis = "columns", inplace = True)
df.head()

#NOTE:
# After inplace it won't repeat previous DATA 
# Updated Df value only displayed @ df.head()
# difference, expense column will not shown after this command execution
```

```python
df.to_csv("updated.csv")

#NOTE:
# we informing intruptor to store updated csv data's in Dataframe 
```

```python
df.head()

#NOTE:
# See here it displays updated version ---> after inplace=True
# Difference, expense --> dropped. Bcz changes[Updated in df]
```

### Take 2 separate dataframe and merge(concat) them in single dataframe  

```python
age = list(np.random.randint(32,50,5))
name = ["TH67","Vijay","Lokesh","LEO","Trisha"]
city = ["chennai","mumbai","chennai","chennai","Delhi"]
df3 = pd.DataFrame({"name":name, "age":age, "location":city} , index = np.arange(1,6))
df3
```

```python
age = list(np.random.randint(32,50,6))
name = ["Arjun","Priyaanand","GVM","Rolex","Sanjaydutt","Mansoor"]
city = ["vaga_border","mumbai","kashmir","mumbai","kashmir","mumbai"]
df4 = pd.DataFrame({"name":name, "age":age, "location":city} , index = np.arange(6,12))
df4
```

```python
pd.concat([df3,df4], axis=1)

#NOTE:
# axis 1 --> differentiate both df3 and df4 (in-separte) 
```

```python
df5=pd.concat([df3,df4], axis=0)
df5

#NOTE:
# axis 0 --> Merge df3 and df4 (Together) 
```

#------------------------------------------------------------------------------------------
# **joins in pandas**
## *1. inner join*

> Comparing (left table --> df5) and (right table --> df6) 

> **Only matching records of "left table" and "right table" will be taken** and considered. (Unmatched --> Eliminated)








```python
df5
```

```python
location = ["chennai", "mumbai", "kashmir"]
fav_food = ["Biryani", "Chocolate", "Chai"]
df6 = pd.DataFrame({"location": location ,"fav_food": fav_food})
df6
```

```python
pd.merge(df5,df6, on = ["location"], how = "inner")
```

## *2. left join*

>  Comparing (left table --> df5) and (right table --> df6) 

>  **All records from "left-table" & matching record of 'right-table'** will *unioned* and considered.



```python
df5
```

```python
df6
```

```python
pd.merge(df5,df6, on = ["location"], how = "left")
```

## *3. right join*

>  Comparing (left table --> df5) and (right table --> df6) 

>  **All records from "right-table" & matching record of 'left-table'** will *unioned* and considered.

```python
df5
```

```python
df6
```

```python
pd.merge(df5,df6, on = ["location"], how = "right")
```

## *4. outer join*

>Comparing (left table --> df5) and (right table --> df6)

>**All records of both "right-table" &  'left-table' will considered** *(Unmatched --> also considered)*

```python
df5
```

```python
df6
```

```python
pd.merge(df5,df6, on = ["location"], how = "outer")
```
