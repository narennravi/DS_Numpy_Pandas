## **Groupby()**

>  groupby() is applicable only for numerics-->[int,float]



```python
df.groupby(["sex"]).sum()

#NOTE:
# "column --> sex" is grouped 
# And sums up numeric columns values --> age, bmi, children, expenses
```

```python
df.groupby(["sex","smoker"]).mean()

#NOTE:
# "column --> sex & smoker" is grouped 
# Provides Mean value of all numeric columns
```

```python
df.groupby(["sex","smoker"])["age","bmi"].mean()

#NOTE:
# "column --> sex & smoker" is grouped 
# Provides "Mean value of column --> age & bmi" only
```

#---------------------------------------------------------------------------------
# TASK: 1

### we aggregating all columns by giving each function(mean,var,sum) to each column[age,bmi,children,expenses]

```python
df.groupby(["sex","smoker"]).agg({"age":"mean", "bmi":"var", "children":"sum", "region":"count", "expenses":"sum"},inplace=False)

#NOTE:
# sex,smoker is grouped
# "age" column gives mean
# "bmi" column gives var
# "children, expenses" column gives sum 
```

# TASK : 1.2
## How to round off age along with aggregation of columns(as previous i/p)

```python
df = df.groupby(["sex","smoker"]).agg({"age":np.mean, "bmi":"var", "children":"sum", "region":"count", "expenses":"sum"}).reset_index()
```

```python
df["age"] = df["age"].round()


#NOTE:
# see now age column is rounded off
# reseted index --> proper index numbering
```

# **TASK : 4**

```python
#if bmi < 25 , normal
#if bmi > 40 , obese
#else bmi is normal

#df => add a column and apply this condition
```

```python
def bmi(x):
  if x < 25:
    return "normal"
  elif x > 40:
    return "obese"
  else:
    return "normal"
```

```python
df["status"] = df["bmi"].apply(bmi)
df

#NOTE:
# "status" column is created ---> display bmi status for df
```

# **TASK: 5**
